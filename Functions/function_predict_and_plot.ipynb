```python
def predict_and_plot(model, filename, class_names):
  """
  Imports an image located at filename, makes a prediction on the image with
  a trained model and plots the image with the predicted class as the title.

  Parameters:
  model (str) = Name of model
  filename (str) = Name of filepath 
  class_names (str) = Names of classes

  """
  # Import the target image, preprocess it
  image = load_and_prep_image(filename)

  # Make a prediction
  predict = model.predict(tf.expand_dims(image, axis=0))

  # Get the predicted class
  if len(predict[0]) > 1: # check for multi-class
    predict_class = class_names[predict.argmax()] # if more than one output, take the max
  else:
    predict_class = class_names[int(tf.round(predict)[0][0])] # if only one output, round

  # Plot the image and predicted class
  plt.imshow(image)
  plt.title(f"Prediction: {predict_class}")
  plt.axis(False);
```
