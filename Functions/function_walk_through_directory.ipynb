```python
import os

def walk_through_directory(directory_path):
  """
  Walks through directory_path returning its contents.

  Parameters:
    directory_path (str): target directory
  
  Returns:
    A print out of:
      number of subdiretories in directory_path
      number of files in each subdirectory
      name of each subdirectory

  """
  for dirpath, dirnames, filenames in os.walk(directory_path):
    print(f"There are {len(dirnames)} directories and {len(filenames)} images in '{dirpath}'.")
```
